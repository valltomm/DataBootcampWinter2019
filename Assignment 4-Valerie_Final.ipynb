```python
#Most Popular Items
#Identify the 5 most popular items by purchase count, then list (in a table):
#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value
```

```python
popular_items = purchase_data_df.groupby(["Item ID","Item Name","Price"])["Price"].agg(["count","sum"])
popular_items = popular_items.rename(columns={"count":"Purchase Count",
                           "sum":"Total Purchase Value",
                           "Price":"Item Price"}).reset_index()
popular_items_new = popular_items.nlargest(5,"Purchase Count")
popular_items_new["Price"] = popular_items_new["Price"].map("${:.2f}".format)
popular_items_new["Total Purchase Value"] = popular_items_new["Total Purchase Value"].map("${:.2f}".format)
popular_items_new
```

```python
#Most Profitable Items
#Identify the 5 most profitable items by total purchase value, then list (in a table):
#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value
```

```python
profit_items = popular_items.nlargest(5, "Total Purchase Value")
profit_items["Price"] = profit_items["Price"].map("${:.2f}".format)
profit_items["Total Purchase Value"] = profit_items["Total Purchase Value"].map("${:.2f}".format)
profit_items
```
